```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.random import sample_without_replacement
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import preprocessing
```

```python
df = pd.read_csv('../Project_4/data/fma_30secs.csv')
df.head()
df.shape
```

```python
# remove null values
df.dropna()
```

```python
df.groupby('label')['filename'].count()
```

```python
# converting string labels into numbers
le = preprocessing.LabelEncoder()
X = df.drop(['label','filename'], axis = 1)
y = le.fit_transform(df.label.values)
```

```python
# Spliting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = .1)
```

```python
# Scaling the X data by using StandardScaler()
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
# Generating the model 
clf =  RandomForestClassifier(n_estimators = 100).fit(X_train_scaled, y_train)
y_pred = clf.predict(X_test_scaled)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
print(f"Training Data Score: {clf.score(X_train_scaled, y_train)}")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
