```python
%matplotlib inline
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#url = 'Resources/fma_3secs.csv'
orignial_df = pd.read_csv(Path('Resources/fma_3secs.csv'))
orignial_df
```

```python
orignial_df.isnull().values.any()
```

```python
orignial_df_subset = orignial_df[(orignial_df.label == 'Instrumental') | (orignial_df.label == 'International')]
orignial_df_subset
```

```python
#drop label and filename columns
X = orignial_df_subset.drop(['label','filename'], axis=1)
# get label values 
y = orignial_df_subset['label']
y.unique()
```

```python
#check if data is balanced
pd.DataFrame(y)['label'].value_counts()
```

```python
X = X[X.columns.drop(list(X.filter(regex='var')))]
X
```

import seaborn as sns
sns.pairplot(X)

```python
print("Shape: ", X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### With unscaled data

```python

classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import classification_report
y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```

### With Scaled Data

```python
# Create a StandardScaler() model and fit it to the training data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
predictions = classifier.predict(X_test_scaled)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import classification_report
y_pred = classifier.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```
